```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
df = pd.read_csv('D:\GenAI\Assignment4thFrb\data\sample.csv')
```

```python
df.head()
```

```python
# One-hot encoding
one_hot_encoding = pd.get_dummies(df['text'])
```

```python
one_hot_encoding
```

```python
# Display the results
print("One-hot encoding:")
print(one_hot_encoding.head())
```

```python
# Bag-of-Words (BOW) with N-grams
# specifing the n-gram range you want, e.g., (1, 1) for unigrams, (2, 2) for bigrams, (1, 2) for unigrams and bigrams
vectorizer = CountVectorizer(ngram_range=(1, 2))  # Change ngram_range as needed
bow_matrix = vectorizer.fit_transform(df['text'])
```

```python
print("\nBag-of-Words (BOW) matrix:")
print(pd.DataFrame(bow_matrix.toarray(), columns=vectorizer.get_feature_names_out()).head())
```

```python
# TF-IDF (Term Frequency-Inverse Document Frequency)
tfidf_vectorizer = TfidfVectorizer(ngram_range=(1, 1))  # Change ngram_range as needed
tfidf_matrix = tfidf_vectorizer.fit_transform(df['text'])
```

```python
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
```

```python
print("\nTF-IDF matrix:")
print(tfidf_df.head())
```


### One-Hot Encoding (OHE):

#### Advantages:
1. Simple Representation: OHE is a straightforward technique that represents each unique word or token as a binary feature.
2. Interpretability: The resulting matrices are easy to interpret, making it clear which words are present or absent.
3. No Weighting: OHE does not assign weights to words, making it useful for scenarios where the importance of words is binary (present or not present).
4. Applicability: OHE can be applied to categorical data beyond text, making it versatile in various contexts.
5. Memory Efficiency: OHE can be memory-efficient for datasets with a limited vocabulary and shorter documents.

#### Disadvantages:
1. High Dimensionality: OHE can lead to high-dimensional matrices, especially in datasets with a large vocabulary, which can be computationally expensive.
2. Loss of Sequence Information: OHE does not capture the sequential or contextual information of words within a document.
3. Sparsity: The resulting matrices are often sparse, leading to increased memory requirements.
4. Equal Importance: All words are treated equally, regardless of their frequency or importance in the corpus.
5. Not Suitable for Continuous Data: OHE is not suitable for scenarios where the order or proximity of words matters.

### Bag-of-Words (BOW):

#### Advantages:
1. Simplicity: BOW is a simple and intuitive representation that disregards word order.
2. Efficiency: BOW is computationally efficient and works well with large datasets.
3. Versatility: BOW can be used in various natural language processing (NLP) tasks, such as sentiment analysis and text classification.
4. Interpretability: The resulting feature vectors are interpretable and can be used to identify important words.
5. Language Independence: BOW is language-independent and can be applied to multiple languages without modification.

#### Disadvantages:
1. Loss of Sequence Information: BOW ignores the order of words, leading to the loss of important sequential information.
2. Sparsity: The resulting matrices can be sparse, especially in the presence of a large vocabulary.
3. Equal Importance: Like OHE, BOW treats all words equally, disregarding their importance in the corpus.
4. Limited Semantics: BOW does not capture the semantic meaning of words, leading to limitations in understanding context.
5. Not Robust to Synonyms: BOW treats synonyms as distinct entities, which can affect performance in tasks requiring synonym recognition.

### N-grams:

#### Advantages:
1. Captures Local Context: N-grams capture local context by considering adjacent word combinations.
2. Improved Semantic Understanding: Compared to BOW, N-grams provide a better understanding of the relationships between neighboring words.
3. Versatility: N-grams can be adjusted to capture different levels of context, from unigrams to higher-order n-grams.
4. Increased Information: Including n-grams in the representation can lead to a richer feature set, improving model performance.
5. Partially Resolves Sparsity: N-grams help address the sparsity issue to some extent by incorporating neighboring word information.

#### Disadvantages:
1. Increased Dimensionality: Higher-order n-grams can result in increased dimensionality, leading to computational challenges.
2. Data Size Dependence: Performance may depend on the size of the dataset, and small datasets may not benefit significantly from n-grams.
3. Computational Complexity: The computational cost of processing and storing higher-order n-grams can be high.
4. Limited Context Understanding: Despite capturing local context, n-grams may still struggle with understanding broader contextual relationships.
5. Loss of Structure: N-grams may not fully capture sentence or document structure, as they focus on local combinations.

### TF-IDF (Term Frequency-Inverse Document Frequency):

#### Advantages:
1. Weighted Representation: TF-IDF assigns weights to words based on their frequency in a document and rarity across the corpus, offering a more nuanced representation.
2. Addresses Sparsity: TF-IDF helps mitigate the sparsity issue by giving higher importance to less frequent words.
3. Semantic Understanding: TF-IDF captures the semantic meaning of words by considering their importance in both a specific document and the entire corpus.
4. Contextual Relevance: Provides a measure of how relevant a word is to a particular document within the broader context of the corpus.
5. Reduces Noise: TF-IDF downweights common words that may not contribute much information, reducing noise in the representation.

#### Disadvantages:
1. Complexity: The TF-IDF algorithm is more complex than simpler techniques like BOW and OHE.
2. Loss of Local Context: Like BOW, TF-IDF may lose local context information, as it considers individual words in isolation.
3. Language Dependence: TF-IDF may be influenced by language-specific characteristics, potentially requiring language-specific adjustments.
4. Not Robust to Synonyms: Similar to BOW, TF-IDF treats synonyms as distinct entities.
5. Parameter Sensitivity: Performance may be sensitive to the choice of parameters, such as the weighting scheme and normalization method.

In summary, the choice between OHE, BOW, N-grams, and TF-IDF depends on the specific requirements and characteristics of your NLP task. Each technique has its strengths and weaknesses, and the most suitable representation often varies based on the nature of the data and the goals of the analysis.
